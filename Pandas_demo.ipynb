```python
country_respondents=df['Country'].value_counts()
country_respondents
```

```python
country_uses_Angular = country_grp['WebframeWorkedWith'].apply(lambda x:x.str.contains('Angular').sum())
country_uses_Angular
```

```python
Angular_df = pd.concat([country_respondents, country_uses_Angular], axis='columns', sort=False)
Angular_df
```

```python
Angular_df.rename(columns={'Country':'NumRespondents', 'WebframeWorkedWith':'NumKnowsAngular'},inplace=True)
Angular_df
```

```python
Angular_df['PctKnowsPython']=(Angular_df['NumKnowsAngular']/Angular_df['NumRespondents'])*100
Angular_df
```

```python
Angular_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```
